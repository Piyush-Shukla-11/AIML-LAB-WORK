```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({'Patient-ID': [11, 12, 13, 14, 15, 16],
                    'Patient-Name': ['Jack', 'Piyush', 'Harshit', 'Anurag', 'Preeti', 'Komal'],
                    'Age': [25, 28, 29, None, 19, 42],
                    'Gender': ['M', 'M', 'M', 'M', 'F', 'F'],
                    'Weight': [75, 80, 95, None, 57, 65],
                    'Diagnosis': ['Fever', 'Dengue', 'Cold', 'Malaria', 'Covid', 'Typhoid'],
                    'Sugar level (mg/dl)': [80, 90, 105, 150, 75, 92],
                    'BP': [120, 130, 140, 150, 88, 100]})
print(df)
```

```python
df.isnull()
```

```python
# HISTOGRAM
plt.hist(df['Age'],bins=5)
plt.ylabel("FREQUENCY")
plt.title("AGE DISTRIBUTION")
plt.xlabel("AGE")
plt.show()
```

```python
# LINE GRAPH
plt.plot(df["Patient-Name"], df["BP"], marker = "o")
plt.xlabel("Patient Name")
plt.ylabel("B.P of Patient")
plt.title("Graph of patient with B.P")
plt.show()
```

```python
# PIE CHART
df["Diagnosis"].value_counts().plot.pie(autopct = "%0.1f%%")
plt.title("Disease distribution")
plt.show()
```
