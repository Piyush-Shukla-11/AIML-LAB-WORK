```python
import numpy as np
a=np.array([[2,13,6],[22,3,5],[7,8,16]])
b=np.array([[15,8,7],[6,16,20],[3,2,1]])
```

```python
print("MATRIX ADDITION\n",a+b)
```

```python
print("MATRIX SUBSTRACTION\n",a-b)
```

```python
print("MATRIX MULTIPLICATION\n",a@b)
```

```python
print("ELEMENT WISE MULTIPLICATION\n",a*b)
```

```python
print("DETERMINANT OF A\n",np.linalg.det(a))
```

```python
print("INVERSE OF A\n",np.linalg.inv(a))
```

```python
print("EIGEN VALUES AND EIGEN VECTORS\n",np.linalg.eig(a))
```

```python
print("MAXIMUM OF A\n",np.max(a))
print("MAXIMUM OF B\n",np.max(b))
```

```python
print("STANDARD VARIATION OF A\n",np.std(a))
```
