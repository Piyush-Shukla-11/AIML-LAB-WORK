```python
# Pandas / Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.DataFrame({
    'Name': ['Jack', 'Joe', 'kris', 'Bob', 'zero'],
    'Age': [19, None, 18, 21, 19],
    'Marks': [85, 78, None, 88, 90],
    'Performance': [5, None, 7, None, 5]})
```

```python
print("No of NAN values:")
print(df.isna().sum())

```

```python
# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Marks'] = df['Marks'].fillna(df['Marks'].median())
df['Performance'] = df['Performance'].fillna(df['Performance'].mean())
print("\n Filled Data:")
print(df)
```

```python
# BAR CHART
sns.barplot(x='Name', y='Marks', data=df)
plt.title("Marks of Students")
plt.show()

```

```python
# BOXPLOT
sns.boxplot(y=df['Marks'])
plt.title("Boxplot of Marks")
plt.show()
```

```python
# PIE CHART
performance_counts = df['Performance'].value_counts()
plt.pie(performance_counts, labels=performance_counts.index, autopct='%1.1f%%')
plt.title("Performance Distribution")
plt.show()
```

```python
#HISTOGRAM
plt.hist(df['Marks'])
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```
